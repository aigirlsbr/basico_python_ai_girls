O método items() retorna uma tupla com o par chave valor

```python
for x,y in voo.items():
    if x == 'London':
        print(y)
    
```

Como o método items() percorre o par chave/valor, a resposta serão dois elementos, por isso preciso de duas variáveis para armazenar. Enquanto x percorre New York, Tokyo, etc, o y percorre no horário, atrasado, etc. Se quero saber só o status do voo, o x percorre as cidades e quando encontra imprime o y, que é o status do voo
